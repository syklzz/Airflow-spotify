```python
import matplotlib.pyplot as plt 
import numpy as np 
```

## Experiment 1

```python
simple_spotify_times = [15, 17, 15, 14, 16, 18, 16, 15, 14, 15]
spotify_times = [27, 29, 30, 28, 26, 27, 26, 28, 31, 27]

N = 10
x = [i for i in range(N)]
spotify_avg = [np.average(spotify_times) for _ in range(N)]
simple_spotify_avg = [np.average(simple_spotify_times) for _ in range(N)]
  
 
plt.plot(x, simple_spotify_times, 'c', label = "Long_DAG", linestyle="-")
plt.plot(x, spotify_times, 'r', label = "Short_DAG", linestyle="-")
plt.plot(x, simple_spotify_avg, 'c', label = "long agv", linestyle="--") 
plt.plot(x, spotify_avg, 'r', label = "short avg", linestyle="--")
plt.legend() 
plt.show()
```

## Experiment 2

```python
short_dag_times = [26, 24, 24, 27, 23, 25, 26, 24, 25, 24]
long_dag_times = [30, 32, 27, 29, 28, 31, 30, 33, 28, 29]


N = 10
x = [i for i in range(N)]
short_dag_avg = [np.average(short_dag_times) for _ in range(N)]
long_dag_avg = [np.average(long_dag_times) for _ in range(N)]
  
 
plt.plot(x, short_dag_times, 'c', label = "Short_DAG", linestyle="-")
plt.plot(x, long_dag_times, 'r', label = "Long_DAG", linestyle="-")
plt.plot(x, short_dag_avg, 'c', label = "short avg", linestyle="--") 
plt.plot(x, long_dag_avg, 'r', label = "long avg", linestyle="--")
plt.legend() 
plt.show()
```
